```python
import numpy as np
import pandas as pd
import os
```

```python
data_dir = '../../data/주어진데이터/유통데이터'
data = os.path.join(data_dir, 'GS리테일_동별 매출지수용 기준값 확인_AMT_NEW.xlsx')
```

### Excel file 설명 (구, 행정코드)

```python
df_table = pd.read_excel(data, header=5, nrows=13, usecols=[1,2,4,5,6], sheet_name='테이블정의서')
```

```python
df_table
```

```python
df_gu_dong = pd.read_excel(data, header=1, nrows=36, usecols=[1,2,3,4], sheet_name='참고)구_행정동코드')
new_col = {'구명':'구','행정동명':'동'}
df_gu_dong = df_gu_dong.rename(columns=new_col)
```

```python
df_gu_dong.head()
```

### 데이터 분석

```python
df = pd.read_excel(data, sheet_name='종합테이블')
```

```python
df.head()
```

```python
new_col = {'OPER_DT':'날짜', 'BOR_CD':'구코드','ADMD_CD':'행정동코드','AMT_IND':'매출지수',
           'LCLS_10_P':'식사_비중','LCLS_20_P':'간식_비중','LCLS_30_P':'마실거리_비중','LCLS_40_P':'홈&리빙_비중',
           'LCLS_50_P':'헬스&뷰티_비중','LCLS_60_P':'취미&여가활동_비중','LCLS_70_P':'사회활동_비중','LCLS_80_P':'임신/육아_비중', 'LCLS_90_P' : '기호품_비중'}
df = df.rename(columns=new_col)
df['날짜'] = pd.to_datetime(df['날짜'], format='%Y%m%d')
```

```python
df.head()
```

```python
df = pd.merge(df, df_gu_dong)
del df['구코드']
del df['행정동코드']
```

```python
df.to_csv('../../data/가공데이터/유통데이터.csv', index=None)
```
