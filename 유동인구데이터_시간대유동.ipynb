```python
import os
import pandas as pd
```

```python
dir_path = '../../data/주어진데이터/유동인구데이터/시간대유동/'
```

```python
csv_list = os.listdir(dir_path)[1::]
```

```python
csv_list
```

```python
for idx, csv in enumerate(csv_list):
    csv = os.path.join(dir_path, csv)
    if idx == 0:
        df = pd.read_csv(csv, delimiter='|')
    else:
        df = df.append(pd.read_csv(csv, delimiter='|'), )
```

```python
del(df['STD_YM'])
```

```python
new_col = {'STD_YMD':'날짜', 'HDONG_NM':'동'}
df = df.rename(columns=new_col)
df['날짜'] = pd.to_datetime(df['날짜'], format='%Y%m%d')
```
